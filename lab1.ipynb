Задание 5.1. Постройте биномиальное распределение для серии из n независимых испытаний с вероятностью успеха р, пуассоновское распределение с параметром λ, гипергеометрическое распределение с параметрами N, М, n, геометрическое распределение с параметрами n, р. Для каждого распределения выполните следующее:

- проверьте равенство Σрk = 1, где рк = P(ξ = k);
- найдите значение к, для которого величина P(ξ = k) максимальна;
- постройте графики распределения и функций распределения;
- вычислите вероятность попадания значений случайной величины в интервал (а, b).

Мой вариант 10 - n=27, p=0.19, λ=0.55, N=190, M=180, a=3, b=8

Биномиальное распределение

![image.png](img/binom.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Параметры
n = 27
p = 0.19

# Вероятности P(X = k)
x = np.arange(0, n + 1)
probabilities = binom.pmf(x, n, p)

# Проверка суммы
print(f"Сумма вероятностей (биномиальное): {np.sum(probabilities)}")

# Поиск максимума
k_max = np.argmax(probabilities)
print(f"k (максимум P(X=k)): {k_max}, P(X={k_max})={probabilities[k_max]}")

# Графики
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, probabilities, color='blue', alpha=0.7)
plt.title("Биномиальное распределение")
plt.xlabel("k")
plt.ylabel("P(X=k)")


plt.tight_layout()
plt.show()

# Вероятность попадания в интервал (3, 8)
a, b = 3, 8
prob_interval = binom.cdf(b, n, p) - binom.cdf(a, n, p)
print(f"Вероятность попадания в интервал (3, 8): {prob_interval}")
```

Пуассоновское распределение

![image.png](img/puasson.png)

```python
from scipy.stats import poisson

# Параметры
lambda_param = 0.55

# Вероятности P(X = k)
k_values = np.arange(0, 20)  # будем рассматривать от 0 до 19
probabilities = poisson.pmf(k_values, lambda_param)

# Проверка суммы
print(f"Сумма вероятностей (пуассоновское): {np.sum(probabilities)}")

# Поиск максимума
k_max = np.argmax(probabilities)
print(f"k (максимум P(X=k)): {k_max}, P(X={k_max})={probabilities[k_max]}")

# Графики
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(k_values, probabilities, color='blue', alpha=0.7)
plt.title("Пуассоновское распределение")
plt.xlabel("k")
plt.ylabel("P(X=k)")


plt.tight_layout()
plt.show()

# Вероятность попадания в интервал (3, 8)
a, b = 3, 8
prob_interval = poisson.cdf(b, lambda_param) - poisson.cdf(a, lambda_param)
print(f"Вероятность попадания в интервал (3, 8): {prob_interval}")
```

Гипергеометрическое распределение

![image.png](img/hypgeom.png)

```python
from scipy.stats import hypergeom

# Параметры
N = 190  # общее количество
M = 180  # количество успешных
n = 27   # количество выборок

# Вероятности P(X = k)
k_values = np.arange(0, min(M, n) + 1)
probabilities = hypergeom.pmf(k_values, N, M, n)

# Проверка суммы
print(f"Сумма вероятностей (гипергеометрическое): {np.sum(probabilities)}")

# Поиск максимума
k_max = np.argmax(probabilities)
print(f"k (максимум P(X=k)): {k_max}, P(X={k_max})={probabilities[k_max]}")

# Графики
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(k_values, probabilities, color='blue', alpha=0.7)
plt.title("Гипергеометрическое распределение")
plt.xlabel("k")
plt.ylabel("P(X=k)")


plt.tight_layout()
plt.show()

# Вероятность попадания в интервал (3, 8)
a, b = 3, 8
prob_interval = hypergeom.cdf(b, N, M, n) - hypergeom.cdf(a, N, M, n)
print(f"Вероятность попадания в интервал (3, 8): {prob_interval}")
```

Геометрическое распределение

![image.png](img/geom.png)

```python
from scipy.stats import geom

# Параметры
p = 0.19

# Вероятности P(X = k)
k_values = np.arange(1, 20)  # k начинается с 1 для геометрического
probabilities = geom.pmf(k_values, p)

# Проверка суммы
print(f"Сумма вероятностей (геометрическое): {np.sum(probabilities)}")

# Поиск максимума
k_max = np.argmax(probabilities)
print(f"k (максимум P(X=k)): {k_max+1}, P(X={k_max+1})={probabilities[k_max]}")

# Графики
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(k_values, probabilities, color='blue', alpha=0.7)
plt.title("Геометрическое распределение")
plt.xlabel("k")
plt.ylabel("P(X=k)")


plt.tight_layout()
plt.show()

# Вероятность попадения в интервал (3, 8)
a, b = 3, 8
prob_interval = geom.cdf(b, p) - geom.cdf(a, p)
print(f"Вероятность попадания в интервал (3, 8): {prob_interval}")
```
