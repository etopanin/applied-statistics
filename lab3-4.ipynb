```python
import numpy as np

data = [
    57.3, 75.1, 78.1, 69.3, 60.1, 77.3, 66.1, 69.5, 72.1, 
    68.7, 81.1, 69.4, 63.1, 67.4, 77.1, 82.6, 64.8, 72.5, 
    62.5, 80.7, 77.6, 65.8, 78.3, 57.7, 80.7, 64.4, 82.8, 
    67.3, 83.1, 70.6, 75.3, 58.0, 60.7, 81.3, 67.1, 69.6, 
    82.4, 62.3, 66.9, 80.6, 62.7, 73.8, 68.9, 83.8, 57.0, 
    72.6, 65.6, 78.7, 59.5, 70.0, 73.5, 58.1, 64.0, 83.9, 
    84.0, 63.5, 74.1, 77.7, 68.5, 80.5, 66.3, 73.0, 79.1, 
    71.1, 80.4, 62.1, 66.7, 83.7, 76.8, 59.3, 71.3, 63.7, 
    71.2, 78.9, 65.2, 77.9, 74.9, 69.1, 70.8, 74.8, 71.6, 
    72.9, 61.9, 71.5, 75.4, 71.7, 59.9, 74.3, 76.1, 70.9, 
    61.3, 71.4, 71.8, 65.0, 67.8, 75.5, 71.9, 64.9, 74.7, 
    62.9
]

# Размер выборки
n = len(data)

# Выборочное математическое ожидание
mean_x = np.mean(data)

# Выборочная дисперсия
D = np.var(data, ddof=1)  # ddof=1 для выборочной дисперсии

print(f'Выборочное математическое ожидание (x̄): {mean_x:.4f}')
print(f'Выборочная дисперсия (D): {D:.4f}')
```

```python
from scipy.stats import chi2, norm

# Определяем интервалы
bins = np.arange(57.0, 84.0 + 3.0, 3.0)
observed_frequencies, _ = np.histogram(data, bins=bins)

# Ожидаемые частоты
mean = mean_x
std_dev = np.sqrt(D)
expected_frequencies = []
total_count = len(data)

for i in range(len(bins) - 1):
    # Вероятность попадания в интервал
    prob = norm.cdf(bins[i+1], loc=mean, scale=std_dev) - norm.cdf(bins[i], loc=mean, scale=std_dev)
    expected_frequencies.append(prob * total_count)

# Вычисляем критерий хи-квадрат
chi_square_statistic = np.sum((observed_frequencies - expected_frequencies) ** 2 / expected_frequencies)

# Общее количество интервалов
k = len(bins) - 1
# Степени свободы
p = 2  # два параметра: среднее и стандартное отклонение
df = k - p - 1

# Критическое значение для уровня значимости α = 0.025
alpha = 0.025
critical_value = chi2.ppf(1 - alpha, df)

print(f'Хи-квадрат: {chi_square_statistic:.4f}')
print(f'Критическое значение: {critical_value:.4f}')

if chi_square_statistic > critical_value:
    print('Нулевая гипотеза отвергается: распределение не нормально.')
else:
    print('Нет оснований отвергать нулевую гипотезу: распределение нормально.')
```

```python
from scipy.stats import t, chi2

# Уровень доверия
gamma = 0.9
alpha = 1 - gamma

# Для математического ожидания
t_critical = t.ppf(1 - alpha / 2, n - 1)
confidence_interval_mean = (mean - t_critical * np.sqrt(D / n), mean + t_critical * np.sqrt(D / n))

# Для стандартного отклонения
chi2_lower = chi2.ppf(alpha / 2, n - 1)
chi2_upper = chi2.ppf(1 - alpha / 2, n - 1)
confidence_interval_std = (
    np.sqrt((n - 1) * D / chi2_upper),
    np.sqrt((n - 1) * D / chi2_lower)
)

print(f'Доверительный интервал для математического ожидания: {confidence_interval_mean}')
print(f'Доверительный интервал для стандартного отклонения: {confidence_interval_std}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Точки (x, y) из таблицы
x_points = [300, 300, 300, 400, 400, 400, 500, 500, 500, 600, 600, 600, 700, 700, 700, 800, 800]
y_points = [0.3, 0.4, 0.5, 0.5, 0.6, 0.7, 0.6, 0.7, 0.8, 0.6, 0.7, 0.8, 0.7, 0.8, 0.9, 0.9, 1.0]

# Диапазон для X
x_range = np.linspace(200, 900, 100)

# Уравнение y = f(x): y = 0.362 + 0.000616 * x
y_regression_1 = 0.362 + 0.000616 * x_range

# Уравнение x = f(y): x = 105.83 + 630.62 * y, выразим y как x
# y = (x - 105.83) / 630.62
y_regression_2 = (x_range - 105.83) / 630.62

# Построение графика
plt.figure(figsize=(8, 6))
plt.scatter(x_points, y_points, color='blue', label='Data points')
plt.plot(x_range, y_regression_1, color='red', label='y = 0.362 + 0.000616x')
plt.plot(x_range, y_regression_2, color='green', label='y = (x - 105.83) / 630.62')
plt.xlabel('X (thousands of monetary units)')
plt.ylabel('Y (tons)')
plt.title('Regression Lines: y = f(x) and x = f(y)')
plt.grid(True)
plt.legend()
```
