Задание 5.6. Постройте графики плотности распределения и функции распределения хи-квадрат с указанным числом степеней свободы, равным k = N. Здесь N - номер варианта.
Задание 5.7. Постройте графики плотности распределения и функции распределения Стьюдента с числом степеней свободы, равным k = N. Здесь N - номер варианта.
Задание 5.8. Постройте графики плотности распределения и функции распределения Фишера для указанных значений n и m.

Мой вариант 10 - β=0.91, ε=0.05, n=5, m=3

![image.png](img/xi_double.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Параметры
k = 10  # степени свободы
x = np.linspace(0, 30, 1000)

# Плотность распределения
pdf = chi2.pdf(x, k)

# Функция распределения
cdf = chi2.cdf(x, k)

# Построение графиков
plt.figure(figsize=(12, 6))

# График плотности
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label=f'chi-squared (k={k})', color='blue')
plt.title('Плотность распределения хи-квадрат')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.grid()
plt.legend()

# График функции распределения
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label=f'chi-squared (k={k})', color='orange')
plt.title('Функция распределения хи-квадрат')
plt.xlabel('x')
plt.ylabel('CDF')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import t

# Параметры
k = 10  # степени свободы
t_values = np.linspace(-5, 5, 1000)

# Плотность распределения
pdf_t = t.pdf(t_values, k)

# Функция распределения
cdf_t = t.cdf(t_values, k)

# Построение графиков
plt.figure(figsize=(12, 6))

# График плотности
plt.subplot(1, 2, 1)
plt.plot(t_values, pdf_t, label=f't-distribution (k={k})', color='blue')
plt.title('Плотность распределения Стьюдента')
plt.xlabel('t')
plt.ylabel('Плотность вероятности')
plt.grid()
plt.legend()

# График функции распределения
plt.subplot(1, 2, 2)
plt.plot(t_values, cdf_t, label=f't-distribution (k={k})', color='orange')
plt.title('Функция распределения Стьюдента')
plt.xlabel('t')
plt.ylabel('CDF')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import f

# Параметры
n = 5  # степени свободы в числителе
m = 3  # степени свободы в знаменателе
x_f = np.linspace(0, 5, 1000)

# Плотность распределения
pdf_f = f.pdf(x_f, n, m)

# Функция распределения
cdf_f = f.cdf(x_f, n, m)

# Построение графиков
plt.figure(figsize=(12, 6))

# График плотности
plt.subplot(1, 2, 1)
plt.plot(x_f, pdf_f, label=f'F-distribution (n={n}, m={m})', color='blue')
plt.title('Плотность распределения Фишера')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.grid()
plt.legend()

# График функции распределения
plt.subplot(1, 2, 2)
plt.plot(x_f, cdf_f, label=f'F-distribution (n={n}, m={m})', color='orange')
plt.title('Функция распределения Фишера')
plt.xlabel('x')
plt.ylabel('CDF')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
